### Visualizations

### Types of flavors and courses

```python
data['flavor_profile'].mode()
```

```python
data['flavor_profile'] = data['flavor_profile'].str.replace('-1', 'spicy')
```

```python
explode = (0.05,0.05,0.05,0.05)
colors = ['silver', 'black', 'crimson', 'steelblue']
plt.figure(figsize=(5,7))
data.groupby(['flavor_profile']).size().plot(kind='pie', ylabel='Flavors of Food', autopct='%1.0f%%', explode=explode, colors=colors)
plt.show()
```

There are total of 5 flavors and spicy is high in percentage 

### Analysis on preparation time and type of diet

```python
sns.histplot(data=data, x='prep_time', hue='diet', element='step', common_norm=True)
```

We can observe from the plot that vegetarain takes more time for preparation

### Analysis on cook time and type of diet

```python
sns.histplot(data=data, x='cook_time', hue='diet', element='step', common_norm=True)
```

We can observe from the plot that vegetrain takes more time for cooking

### Type of Diet on region and state

```python
sns.catplot(x='region', y='state', hue='diet', data=data, aspect=1.5)
plt.show()
```

We can observe from he plot that there are more vegetarian diets than non-vegetarian

```python
sns.catplot(x='region', y='state', hue='course', data=data, aspect=1.5)
plt.show()
```

With the catplot, we can conclude that people prefer main course than other courses
