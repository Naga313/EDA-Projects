### Visulaizations

### Type of sectors

```python
plt.figure(figsize=(20,20))
data.groupby('Sector').size().plot(kind='pie', autopct='%0.0f%%', ylabel='Sectors')
plt.show()
```

We can observe from the plot there are wide variety of sectors

### Comparision of country and CAGR rate in the year 2020 

```python
plt.figure(figsize=(10,5))
sns.barplot(data=data, x='Country', y='CAGR', hue='Ranked2020')
plt.xticks(rotation=90)
plt.show()
```

We can observe from the plot that countries like Denmark and Estonia have a good range of CAGR and ranked in 2020

### Comparision of country and CAGR rate in the year 2021

```python
plt.figure(figsize=(10,5))
sns.barplot(data=data, x='Country', y='CAGR', hue='Ranked2021')
plt.xticks(rotation=90)
plt.show()
```

We can observe from the plot that Lativa tops high CAGR and ranked in 2021

### What is the revenue of the companies w.r.t employees in the year 2017

```python
plt.figure(figsize=(10,5))
plt.scatter(x='Revenue2017', y='Employees2017', data=data)
plt.xlabel('Revenue in 2017')
plt.ylabel('Employees in 2017')
plt.show()
```

We can observe from the plot that revenue lies between 0-0.6 with approximately 500 employees, there are ouliers also between(0.8-1.4)

### What is the revenue of the companies w.r.t employees in the year 2020

```python
plt.figure(figsize=(10,5))
plt.scatter(x='Revenue2020', y='Employees2020', data=data)
plt.xlabel('Revenue in 2020')
plt.ylabel('Employees in 2020')
plt.show()
```

We can observe from the plot that revenue lies between 0-0.5 with approximately 1250 employees, there are ouliers also between(0.5-2.0)

With the scatterplot we can conclude that there is a increrase of employees of 2.2x times and revenue of 0.15x times from the year 2017 to 2020
