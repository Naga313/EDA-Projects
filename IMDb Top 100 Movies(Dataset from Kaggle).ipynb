## Visualizations

### What is the highest rating?

```python
sns.histplot(data=data, x="imdb_rating", kde=True)
```

We can observe that the average rating of top 100 movies is 8.35

### What is the relation between category and imdb rating?

```python
sns.boxplot(x='imdb_rating', y='category', data=data)
```

For category the average imdb_rating is 8.0 to 8.6

### What is the relation between votes and imdb rating?

```python
plt.figure(figsize=(25,10))
plt.scatter(data=data, x='votes', y='imdb_rating', c='black', marker='*', edgecolor='blue', s=200)
plt.xlabel('Votes')
plt.ylabel('IMDb rating')
plt.xticks(rotation=90)
plt.show()
```

We can observe from the plot that there is decrease in votes as rating decreases 

### Runtime in minutes

```python
data['run_time'] = data['run_time'].str.replace('min',' ',regex=True)
```

```python
plt.figure(figsize=(15,6))
sns.histplot(data.run_time, bins=5)
plt.xticks(rotation=90)
plt.show()
```

Most of the movies have run time of 127 minutes

### What is the relation between run time and votes?

```python
plt.figure(figsize=(20,15))
plt.scatter(data=data, x='votes', y='run_time', c='orange', marker='o', edgecolor='red', s=100)
plt.xlabel('Votes')
plt.ylabel('IMDb rating')
plt.xticks(rotation=90)
plt.show()
```

With the scater plot, we can conclude that higher the votes the duration will be more and movie is intresting to watch
